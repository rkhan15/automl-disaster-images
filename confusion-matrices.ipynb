```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

# AWS Rekognition results (from API)

```python
#                     PREDICTED
#             no-damage   destroyed
# TRUE no-damage   456         20
# TRUE destroyed   14          93
```

```python
aws_cf = pd.DataFrame(columns=['no-damage', 'destroyed'], index=['no-damage', 'destroyed'])
aws_cf.loc['no-damage']['no-damage'] = 456
aws_cf.loc['no-damage']['destroyed'] = 20
aws_cf.loc['destroyed']['no-damage'] = 14
aws_cf.loc['destroyed']['destroyed'] = 93
aws_cf['no-damage'] = aws_cf['no-damage'].astype(int) 
aws_cf['destroyed'] = aws_cf['destroyed'].astype(int) 
aws_cf
```

```python
aws_cf_pct = aws_cf.copy()
aws_cf_pct['no-damage'] = aws_cf_pct['no-damage'].astype(object) 
aws_cf_pct['destroyed'] = aws_cf_pct['destroyed'].astype(object) 
aws_cf_pct.loc['no-damage']['no-damage'] = f"{round(aws_cf.loc['no-damage']['no-damage'] / (aws_cf.loc['no-damage']['no-damage'] + aws_cf.loc['no-damage']['destroyed']) * 100, 1)}%"
aws_cf_pct.loc['no-damage']['destroyed'] = f"{round(aws_cf.loc['no-damage']['destroyed'] / (aws_cf.loc['no-damage']['no-damage'] + aws_cf.loc['no-damage']['destroyed']) * 100, 1)}%"
aws_cf_pct.loc['destroyed']['no-damage'] = f"{round(aws_cf.loc['destroyed']['no-damage'] / (aws_cf.loc['destroyed']['no-damage'] + aws_cf.loc['destroyed']['destroyed']) * 100, 1)}%"
aws_cf_pct.loc['destroyed']['destroyed'] = f"{round(aws_cf.loc['destroyed']['destroyed'] / (aws_cf.loc['destroyed']['no-damage'] + aws_cf.loc['destroyed']['destroyed']) * 100, 1)}%"

aws_cf_pct
```

```python
aws_cf_pct = aws_cf.copy()
aws_cf_pct['no-damage'] = aws_cf_pct['no-damage'].astype(float) 
aws_cf_pct['destroyed'] = aws_cf_pct['destroyed'].astype(float) 
aws_cf_pct.loc['no-damage']['no-damage'] = round(aws_cf.loc['no-damage']['no-damage'] / (aws_cf.loc['no-damage']['no-damage'] + aws_cf.loc['no-damage']['destroyed']) * 100, 1)
aws_cf_pct.loc['no-damage']['destroyed'] = round(aws_cf.loc['no-damage']['destroyed'] / (aws_cf.loc['no-damage']['no-damage'] + aws_cf.loc['no-damage']['destroyed']) * 100, 1)
aws_cf_pct.loc['destroyed']['no-damage'] = round(aws_cf.loc['destroyed']['no-damage'] / (aws_cf.loc['destroyed']['no-damage'] + aws_cf.loc['destroyed']['destroyed']) * 100, 1)
aws_cf_pct.loc['destroyed']['destroyed'] = round(aws_cf.loc['destroyed']['destroyed'] / (aws_cf.loc['destroyed']['no-damage'] + aws_cf.loc['destroyed']['destroyed']) * 100, 1)

aws_cf_pct
```

```python
plt.figure(figsize=(6,6))
sns.set(font_scale = 2)
sns.heatmap(aws_cf, square=True, fmt='g', annot=True, cmap='Blues', cbar=False)
plt.title('AWS Rekognition Model Results')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.set(font_scale = 2)
sns.heatmap(aws_cf_pct, square=True, annot=True, cmap='Blues', cbar=False, fmt='.1f')
plt.title('AWS Rekognition Model Results')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
